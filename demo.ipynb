```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the image
gpus = ['1080TI', '2080TI', 'A5000', 'A100', 'RTX 4090', 'H100']
bfloat16 = [0.9, 1.1, 3.45, 2.95, 6.2, 6.8]
bit8 = [0.85, 2.2, 1.2, 0.75, 2.1, 0]
bit4 = [0.92, 1.1, 2.95, 2.2, 6, 4.4]

# Bar width
bar_width = 0.2

# X-axis positions for each group of bars
r1 = np.arange(len(gpus))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width * 2 for x in r1]

# Create the plot with wider width and shorter height
plt.figure(figsize=(12, 4))  # Adjusted figure size for a flatter chart
plt.bar(r1, bfloat16, color='#D3D3D3', width=bar_width, edgecolor='black', label='bfloat16')  # Lighter gray
plt.bar(r2, bit8, color='#ADD8E6', width=bar_width, edgecolor='black', label='8 bit')  # Lighter blue
plt.bar(r3, bit4, color='#90EE90', width=bar_width, edgecolor='black', label='4 bit')  # Lighter green

# Add labels and title
plt.xlabel('GPU Model', fontweight='bold')
plt.ylabel('Actions / sec', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(gpus))], gpus)

# Add legend
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```
